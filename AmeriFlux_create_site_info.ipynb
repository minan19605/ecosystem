```python
import os
import glob
import numpy as np
import pandas as pd
import json
import datetime
import math

# importing required modules
from zipfile import ZipFile
```

```python
zip_dir = 'D:/AmeriFlux/'
all_zipfiles = glob.glob(zip_dir + "*.zip")
```

```python
site_info = []
for zipfile in all_zipfiles:
    with ZipFile(zipfile, 'r') as zip:
        file_list = zip.namelist()
        if len(file_list) > 1:
            for _file in file_list:
                name_split = _file.split('.')
                
                if name_split[1] == 'xlsx':
                    #pre_fix = name_split[0]
                    #site_name = pre_fix.split('_')[1]
                    df = pd.read_excel(zip.open(_file))
                    _data = dict()
                    _data['site_id'] = df['SITE_ID'][0]
                    for _idx, row in df.iterrows():
#                         if row['VARIABLE'] == 'MAT':
#                             _data['MAT'] = row['DATAVALUE']
#                         if row['VARIABLE'] == 'MAP':
#                             _data['MAP'] = row['DATAVALUE']
                        if row['VARIABLE'] == 'LOCATION_LAT':
                            _data['Latitude'] = row['DATAVALUE']
                        if row['VARIABLE'] == 'LOCATION_LONG':
                            _data['Longitude'] = row['DATAVALUE']
                        if row['VARIABLE'] == 'LOCATION_ELEV':
                            _data['Elev'] = row['DATAVALUE']
                        if row['VARIABLE'] == 'UTC_OFFSET':
                            _data['UTC'] = row['DATAVALUE']
                            
    site_info.append(_data)

pd_siteinfo = pd.DataFrame(site_info)
```
