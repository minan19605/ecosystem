```python
def process_values(qcv, era, f_handler):
    msk = np.isfinite(qcv_ta)
    x = qcv[msk]
    y = era[msk]
    if x.size == 0 or y.size == 0:
        s,t,r,p,e = 0,0,0,0,0
    else:
        s,t,r,p,e = linregress(x,y)
    rmse = np.sqrt(np.mean((y-x)**2))
    ratio = msk.sum()/msk.size * 100
    output_txt = '{},{},{},{},{}\n'.format(ratio,s,t,rmse,r)
    f_handler.write(output_txt)
```

```python
qcv_droplist = []
era_droplist = []
for sub_dir in sub_dir_list[1]:
    site_name = sub_dir
    if site_name in _bad:
        continue
    if site_name in _good:
        continue
    _bad.append(site_name)
    full_qcv_dir = src_dir + site_name + qcv_dir
    full_era_dir = src_dir + site_name + era_dir
    qcv_file_list = glob.glob(full_qcv_dir + '*.csv')
    qcv_list = []
    era_list = []
    for qcv_file in qcv_file_list:
        qcv_name = os.path.basename(qcv_file)
        year = qcv_name.split('.')[0][-4:]
        if int(year) > 2021:
            break
        
        era_file = site_name + '_' + year + '.csv'
        era_file = full_era_dir + era_file
        
        df_qcv = pd.read_csv(qcv_file,dtype={'TIMESTAMP_START':str, 'TIMESTAMP_END':str},skiprows=9)
        #df_qcv.replace(-9999.0, np.nan, inplace=True)
        df_era = pd.read_csv(era_file)
#         qcv_size = df_qcv[df_qcv.columns[0]].size 
#         era_size = df_era[df_era.columns[0]].size
#         drop_count = qcv_size - era_size
#         # year 2021, era will missing last few rows for US sites
#         # era original files are using UTC time,and 202112312330 will change to local time
#         if drop_count > 0:
#             df_qcv.drop(df_qcv.tail(drop_count).index, inplace = True)
#             qcv_droplist.append(qcv_name)
#         elif drop_count < 0:
#             df_era.drop(df_era.tail(-drop_count).index, inplace = True)
#             era_droplist.append(era_file)
#         else:
#             # do nothing
#             pass
        
        qcv_list.append(df_qcv)
        era_list.append(df_era)
        
    qcv_full = pd.concat(qcv_list)
    #qcv_full = qcv_full.to_numpy()
    era_full = pd.concat(era_list)
    #era_full = era_full.to_numpy()
    
    txt_file = full_era_dir + 'stat30_' + site_name + '.txt'
    with open(txt_file, 'w') as file:
        file.write('Var,Unit,Percentage of Gaps,Slope,Intercept,RMSE,Corr\n')
        file.write('Ta,Celsius,')
        
        qcv_ta = qcv_full['TA'].replace(-9999.0, np.nan)
        era_ta = era_full['Ta_era']        
        process_values(qcv_ta, era_ta, file)
        
        file.write('Pa,kPa,')
        qcv_pa = qcv_full['PA'].replace(-9999.0, np.nan)
        era_pa = era_full['Pa_era']
        process_values(qcv_pa, era_pa, file)
        
        file.write('VPD,hPa,')
        qcv_vpd = qcv_full['VPD'].replace(-9999.0, np.nan)
        era_vpd = era_full['VPD_era']
        process_values(qcv_vpd, era_vpd, file)
        
        file.write('WS,m s-1,')
        qcv_ws = qcv_full['WS'].replace(-9999.0, np.nan)
        era_ws = era_full['WS_era']
        process_values(qcv_ws, era_ws, file)
        
        file.write('Precip,mm per 30min,')
        qcv_p = qcv_full['P'].replace(-9999.0, np.nan)
        era_p = era_full['Precip_era']
        process_values(qcv_p, era_p, file)
        
        file.write('Rg,W m-2,')
        qcv_sw = qcv_full['SW_IN'].replace(-9999.0, np.nan)
        era_sw = era_full['Rg_era']
        process_values(qcv_sw, era_sw, file)
        
        file.write('LWin,W m-2,')
        qcv_lw = qcv_full['LW_IN'].replace(-9999.0, np.nan)
        era_lw = era_full['LWin_era']
        process_values(qcv_lw, era_lw, file)
        
        file.write('LWin_calc,W m-2,')
        process_values(qcv_lw, era_lw, file)
        
    print("    Output {}".format(txt_file))
    _bad.remove(site_name)
    _good.append(site_name)
```
